```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el DataSetFinal.csv
# Asegúrate de que el archivo 'DataSetFinal.csv' esté en la misma ruta o proporciona la ruta completa.
file_path = 'C:/Users/juand/Downloads/Proyecto_analisis/DataSetFinal.csv'
df_pd = pd.read_csv(file_path)

# Visualizar las primeras filas para revisar los datos
print("Primeras 5 filas del DataSetFinal.csv:")
print(df_pd.head())

# Revisar tipos de columnas e información general
print("\nInformación general del DataSetFinal.csv:")
df_pd.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Distribución de Atributos de los Equipos', fontsize=16)

sns.histplot(df_pd['home_speed'], kde=True, bins=15, ax=axes[0, 0])
axes[0, 0].set_title('Velocidad del Equipo Local (home_speed)')
axes[0, 0].set_xlabel('Velocidad')
axes[0, 0].set_ylabel('Frecuencia')

sns.histplot(df_pd['away_speed'], kde=True, bins=15, ax=axes[0, 1])
axes[0, 1].set_title('Velocidad del Equipo Visitante (away_speed)')
axes[0, 1].set_xlabel('Velocidad')
axes[0, 1].set_ylabel('Frecuencia')

sns.histplot(df_pd['home_aggression'], kde=True, bins=15, ax=axes[1, 0])
axes[1, 0].set_title('Agresividad del Equipo Local (home_aggression)')
axes[1, 0].set_xlabel('Agresividad')
axes[1, 0].set_ylabel('Frecuencia')

sns.histplot(df_pd['away_aggression'], kde=True, bins=15, ax=axes[1, 1])
axes[1, 1].set_title('Agresividad del Equipo Visitante (away_aggression)')
axes[1, 1].set_xlabel('Agresividad')
axes[1, 1].set_ylabel('Frecuencia')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle('Distribución de las Cuotas de Apuestas (B365)', fontsize=16)

sns.histplot(df_pd['B365H'], kde=True, bins=20, ax=axes[0])
axes[0].set_title('Cuotas para Victoria Local (B365H)')
axes[0].set_xlabel('Cuota')
axes[0].set_ylabel('Frecuencia')

sns.histplot(df_pd['B365D'], kde=True, bins=20, ax=axes[1])
axes[1].set_title('Cuotas para Empate (B365D)')
axes[1].set_xlabel('Cuota')
axes[1].set_ylabel('Frecuencia')

sns.histplot(df_pd['B365A'], kde=True, bins=20, ax=axes[2])
axes[2].set_title('Cuotas para Victoria Visitante (B365A)')
axes[2].set_xlabel('Cuota')
axes[2].set_ylabel('Frecuencia')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Distribución de Goles Marcados', fontsize=16)

sns.histplot(df_pd['home_team_goal'], discrete=True, kde=False, ax=axes[0])
axes[0].set_title('Goles del Equipo Local (home_team_goal)')
axes[0].set_xlabel('Número de Goles')
axes[0].set_ylabel('Frecuencia')
axes[0].set_xticks(range(int(df_pd['home_team_goal'].min()), int(df_pd['home_team_goal'].max()) + 1))

sns.histplot(df_pd['away_team_goal'], discrete=True, kde=False, ax=axes[1])
axes[1].set_title('Goles del Equipo Visitante (away_team_goal)')
axes[1].set_xlabel('Número de Goles')
axes[1].set_ylabel('Frecuencia')
axes[1].set_xticks(range(int(df_pd['away_team_goal'].min()), int(df_pd['away_team_goal'].max()) + 1))

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Atributos de los Equipos vs. Resultado del Partido', fontsize=16)

sns.boxplot(data=df_pd, x='resultado', y='home_speed', palette='Pastel2', ax=axes[0])
axes[0].set_title('Velocidad Local vs. Resultado')
axes[0].set_xlabel('Resultado')
axes[0].set_ylabel('Velocidad del Equipo Local')

sns.boxplot(data=df_pd, x='resultado', y='away_speed', palette='Pastel2', ax=axes[1])
axes[1].set_title('Velocidad Visitante vs. Resultado')
axes[1].set_xlabel('Resultado')
axes[1].set_ylabel('Velocidad del Equipo Visitante')

sns.boxplot(data=df_pd, x='resultado', y='away_aggression', palette='Pastel2', ax=axes[2])
axes[2].set_title('Agresividad Visitante vs. Resultado')
axes[2].set_xlabel('Resultado')
axes[2].set_ylabel('Agresividad Visitante')

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.violinplot(data=df_pd, x='resultado', y='home_aggression', palette='Pastel1')
plt.title('Distribución de Agresividad Local vs. Resultado del Partido (Violin Plot)')
plt.xlabel('Resultado')
plt.ylabel('Agresividad del Equipo Local')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(data=df_pd, x='resultado', y='B365H', palette='Spectral')
plt.title('Distribución de Cuotas Locales (B365H) vs. Resultado del Partido (Violin Plot)')
plt.xlabel('Resultado')
plt.ylabel('Cuota B365H')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.regplot(data=df_pd, x='home_speed', y='home_team_goal', scatter_kws={'alpha':0.3})
plt.title('Relación entre Velocidad Local y Goles Anotados (con Regresión)')
plt.xlabel('Velocidad del Equipo Local')
plt.ylabel('Goles del Equipo Local')
plt.show()

plt.figure(figsize=(8, 6))
sns.regplot(data=df_pd, x='home_aggression', y='home_team_goal', scatter_kws={'alpha':0.3}, color='g')
plt.title('Relación entre Agresividad Local y Goles Anotados (con Regresión)')
plt.xlabel('Agresividad del Equipo Local')
plt.ylabel('Goles del Equipo Local')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.jointplot(data=df_pd, x='home_speed', y='home_aggression', kind='scatter', height=7)
plt.suptitle('Distribución Conjunta de Velocidad y Agresividad Local', y=1.02, fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()

sns.jointplot(data=df_pd, x='home_speed', y='home_team_goal', kind='kde', fill=True, height=7, cmap="Blues")
plt.suptitle('Distribución Conjunta (KDE) de Velocidad Local y Goles Locales', y=1.02, fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```
